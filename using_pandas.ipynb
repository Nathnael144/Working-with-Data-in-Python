A Series is a one-dimensional labeled array in Pandas. It can be thought of as a single column of data with labels or indices for each element. 

You can create a Series from various data sources, such as lists, NumPy arrays, or dictionaries
Here's a basic example of creating a Series in Pandas:

```python
import pandas as pd
# Create a Series from a list
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
print(s)
```

```python
print(s[2])
print(s.iloc[1])
print(s[1:4])
```

What is a DataFrames?
A DataFrame is a two-dimensional labeled data structure with columns of potentially different data types. Think of it as a table where each column represents a variable, and each row represents an observation or data point. DataFrames are suitable for a wide range of data, including structured data from CSV files, Excel spreadsheets, SQL databases, and more.

```python
import pandas as pd
# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)
```

```python
print(df['Name'])
```

```python
print(df['Age'] > 28)
```

```python
print(df.iloc[2])   # Access the third row by position
  # Access the second row by label
```

```python
import pandas as pd

# Convert the table to a dictionary
data = {
    'OrderID': [1, 2, 3, 4, 5],
    'Product': ['Laptop', 'Smartphone', 'Desk Chair', 'Notebook', 'Monitor'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Stationery', 'Electronics'],
    'Quantity': [2, 3, 5, 10, 1],
    'Price': [800, 600, 150, 2, 300],
    'Total': [1600, 1800, 750, 20, 300],
    'OrderDate': ['2022-01-10', '2022-02-15', '2022-03-12', '2022-04-05', '2022-05-21'],
    'CustomerCity': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']
}

# Create DataFrame from the dictionary
df_orders = pd.DataFrame(data)
print(df_orders)
```

We can access the column Quantity and assign it a new dataframe x:

```python
x= df_orders[['Quantity']]
print(x)
type(x)
```

```python
y = df_orders[['Product','Category', 'Quantity']]
print(y)
```

```python
y=df_orders.iloc[3,1]
print(y)  

x= df_orders.loc[2,'Quantity']
print(x)
```

```python
x=df_orders.iloc[0:2, 0:3]
print(x)
```

```python
y=df_orders.loc[0:2, 'OrderID':'Category']
print(y)
```
